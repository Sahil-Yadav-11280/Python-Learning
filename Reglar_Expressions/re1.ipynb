#### Metacharacters
unique meaning to regex matching engine


##### Character class ([]) -> matches any single character that is in the class or square brackets
ex- to search any 3 consecutive decimal digits in a string:

```python
s = "foo123bar"
re.search('[0-9][0-9][0-9]',s)
```

```python
print(re.search('[0-9][0-9][0-9]','foo456bar'))
print(re.search('[0-9][0-9][0-9]','234baz'))
print(re.search('[0-9][0-9][0-9]','qux678'))
print(re.search('[0-9][0-9][0-9]', '12foo34'))
print(re.search('ba[artz]', 'foobarqux'))
print(re.search('ba[artz]', 'foobazqux'))
print(re.search('[a-z]', 'FOObar'))

# [0-9a-fA-F] matches any hexadecimal
print(re.search("[0-9a-fA-F]" , "--- a0 ---"))
```

```python
# You can complement a character class by specifying ^ as the first character, in which case it matches any character that isn’t in the set. In the following example, [^0-9] matches any character that isn’t a digit
print(re.search("[^0-9]" , "12345foo"))

# if ^ appears in a character class but isnt the first character then it means nothing and matches a literal '^' character
print(re.search('[0-9^][a-z]' , "oo3^cde"))
print(re.search('[#:^]' , "oo3^cde"))
```

```python
# we use hyphen to set a range of characters in character class but what if we want to search a literal hyphen
# we can do this by putting hyphen in first , last or escape it with a backslash(\)
s2 = "123-456"
print(re.search('[-abc]' , s2))
print(re.search('[abc-]' , s2))
print(re.search(r'[ab\-c]' , s2)) # using rawstring since python doesnt consider \- as a valid escape sequence
```

```python
# If we want to search a literal [ or ] we can either place it as first character or escape it
print(re.search(r'[[abc]',"foo[1]"))
print(re.search(r'[]abc]',"foo[1]"))
print(re.search(r'[]]',"foo[1]"))
print(re.search(r'[\]abc]',"foo[1]"))
```

```python
# other regex metacharacters loose their speacial meanings inside a character class and are treated as literal characters
print(re.search('[)*+|]' , "123*456"))
print(re.search('[)*+|]' , "123+456"))
```

##### Dot metacharacter (.)
1. matches any character except newline
2. also called as wildcard

```python
s="foo123bar"
re.search("1.3",s)
# 1.3 search found bcoz any character except newline is detected by dot metacharacter, here that character was 2
```

```python
print(re.search("1.3","foo13bar"))
# 1.3 didnt matched bcoz here is no character between 1 and 3 that can be searched by dot metacharacter
```

\w matches any alphaumeric character and underscore( _ ) which means it is horthand for [ a-zA-Z0-9_ ]

\W is opposite of it , it matches everything except alphanumeric characters and underscore that is [ ^a-zA-Z0-9_ ]

```python
print(re.search(r"\w" , "#(.a$@&"))
print(re.search("[a-zA-Z0-9_]" , "#(.a$@&"))
```

\d matches any decimal character

\D is opposite , it matches any non decimal caracter
