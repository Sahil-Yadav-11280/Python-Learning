### Tuples
1. Tuples are ordered sequence
2. Comma separated value within parenthesis


```python
ratings = (10,9,6,9,8,10,9)

# can also store value of different datatypes or tuples itself
tuple1 = ("disco" , 5 , 6.9 , (3,2))

# Indexing:
# positive indexing -> 0,1,2,....
# negative indexing -> -1,-2,-3,....
# ex->
tuple1[0]
```

```python
print(tuple1[-1])
print(tuple1[-1][0])

# slicing
print(tuple1[0:2])
print(tuple1[::2])
print(tuple1[::-2])
```

```python
# concatenating
tuple2 = tuple1 + ("hardrock" , 10)
print(tuple2)
print(tuple2[0:3])

# length function
print(len(tuple1))
print(len(tuple2))
```

```python
# Tuples are immutable:
ratings = (10,9,6,9,8,10,9)
ratings1 = ratings

ratings1[0] = 11
```

```python
# However we can concatenate a tuple or create a new tuple if we want to modify it

# ex -> 
Ratings_Sorted = sorted(ratings)
print(Ratings_Sorted)
```

### Lists
1. Lists are also ordered sequences
2. Lists are enclosed in square brackets -> []
3. Lists are mutable

```python
# lists can also store complex elements similar to tuples:
L = ["michael Jackson" , 10.1 , 1982 , [5,"Hello"] , ("tuple in list" , 0.3)]

# Indexing works same as tuples:
print(L[0])
print(L[0:3])
```

```python
# advanced indexing
print(L[3:5])
print(L[3:5][0])
print(L[3:5][0][1])
```

```python
# Extend method:
L = ["michael Jackson" , 10.1 , 1982 , [5,"Hello"] , ("tuple in list" , 0.3)]
L.extend(["pop",10])
print(L) # we just modified L meaning L is mutable
```

```python
# Append method:
L = ["michael Jackson" , 10.1 , 1982 , [5,"Hello"] , ("tuple in list" , 0.3)]
L.append(["pop",10])
print(L)
```

```python
# watching above examples , the diff between extend and append must be understood

# del method:
A = ["hard rock" , 10 ,1.2]
print(A)
del(A[0])
print(A)
```

```python
# Split method:
B = "hard rock".split()
print(B)

# split can also be used to split by setting a custom delimiter (default -> space)
C = "new,old,style,random,text".split(",")
print(C)
```

```python
# Modifiable:
A = ["hard rock" , 10 ,1.2]
B = A

print(B)
print(A)

# now both var A and B point to a single list , any modification on B will show effect on A

B[0] = "soft rock"
print(B)
print(A)
```
