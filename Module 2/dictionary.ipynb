#### Dictionaries
1. Dictionary is similar to a list except here we use keys instead of indexes
2. Index must only be integers starting from 0, however keys can be of any type or any value
3. Dictionaries are enclosed in curly braces

<img src="./images/dict1.png">

```python
Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), ('key5'): 5, (0, 1): 6}
Dict
```

```python
# we can access values using keys similar to index
print(Dict["key1"]) # type: ignore
print(Dict[(0,1)]) # type: ignore
```

```python
# Create a sample dictionary

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

```python
print(release_year_dict["Thriller"])

# get all keys
print(release_year_dict.keys())

# get all values
print(release_year_dict.values())

# get both keys and values
print(release_year_dict.items())
```

```python
# appending
release_year_dict["Graduation"] = "2008"

# Dictionary values are mutable , they can also store duplicate values but keys are neither duplicable nor mutable
release_year_dict["Graduation"] = "2007"

release_year_dict
```

```python
# del function -> deletes an element
del(release_year_dict["Graduation"])
release_year_dict
```

```python
# in keyword:
# To check if an element is present in a dictionary or not
print("Rumours" in release_year_dict)
```
