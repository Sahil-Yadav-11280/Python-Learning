```python
class vehicle:
    color = "white"

    def __init__(lord_sahil , max_speed , mileage):
        lord_sahil.max_speed = max_speed
        lord_sahil.mileage = mileage
        lord_sahil.seat_capacity = None
    
    def assign_seating_capacity(lord_sahil , seating_capacity):
        lord_sahil.seat_capacity = seating_capacity

    def show(lord_sahil):
        print(f"Maximum Speed: {lord_sahil.max_speed}\nAvgerage Mileage: {lord_sahil.mileage}\nSeating Capacity: {lord_sahil.seat_capacity}\nColor: {lord_sahil.color}\n\n\n")
```
